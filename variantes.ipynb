<div style="text-align: center">
    <h2>Seleção de Homólogo para Análise de Variantes</i></h2>
</div>

Com base na proximidade filogenética inferida a partir dos alinhamentos múltiplos e da análise filogenética prévia, a escolha do fago de referência foi orientada pela necessidade de avaliar, com maior precisão, alterações evolutivas recentes e específicas que contribuem para a singularidade do *Staphylococcus phage 88*. A comparação com o homólogo geneticamente mais próximo permite minimizar o ruído associado a divergências antigas e isolar variantes pontuais com potencial impacto funcional.

Nesse contexto, o fago __NC_010147__ (*Staphylococcus phage phiMR11*) foi selecionado como referência, uma vez que apresentou, de forma consistente, os maiores níveis de identidade (>98%) em todas as proteínas analisadas (Holina, Endolisina e Fibra Caudal). Esta elevada similaridade genética possibilita a identificação de diferenças subtis e a inferência mais robusta do seu impacto biológico.
